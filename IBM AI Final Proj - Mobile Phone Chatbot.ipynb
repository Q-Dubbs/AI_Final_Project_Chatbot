```python
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer
from chatterbot.trainers import ListTrainer
```

```python
chatbot = ChatBot("PhoneBot",
                  preprocessors = [
                                   "chatterbot.preprocessors.clean_whitespace",
                                   "chatterbot.preprocessors.unescape_html",
                                   "chatterbot.preprocessors.convert_to_ascii"
                                    ],
                  storage_adapter= "chatterbot.storage.SQLStorageAdapter",
                  logic_adapters = [
                                    {
                                        'import_path': 'chatterbot.logic.BestMatch',
                                        'default_response': 'I am sorry, but I do not understand. I am still learning.',
                                        'maximum_similarity_threshold': 0.90
                                     }
                                    ],
                  database_uri= "sqlite:///content/drive/MyDrive/IBM AI Group 4/PYTHON_CHATBOT/Phonebot_database.sqlite3",
                  )
```

```python
corpus_trainer = ChatterBotCorpusTrainer(chatbot)
corpus_trainer.train("/content/drive/MyDrive/IBM AI Group 4/PYTHON_CHATBOT/Training Data/phones")

# trainer = ListTrainer(chatbot)

# trainer.train(conv)
```

```python
start_bot = input("Type 'y' to start chatbot, type 'n' to exit: ").lower()

if start_bot == "y":
    name_input = input("Enter your name to start chatting: ")
    print("Chatbot started.  Type 'bye' when you want to exit chatbot.  Have fun!")

    start_chat = True
    while start_chat:
        user_input = input(f"{name_input}: ")
        if user_input.lower() == "bye" or "bye bye":
            start_chat = False
            print("PhoneBot: Bye bye. Have a nice day!")
        else:
            bot_input = chatbot.get_response(user_input)
            print(f"PhoneBot: {bot_input}"
            )

else:
    print("Bye Bye!")
```
